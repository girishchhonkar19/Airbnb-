```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
df.describe()
```

## Filling NA values

```python
df.reviews_per_month.fillna(0, inplace=True)
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Plotting Room Type on Neighbourhood group


```python
plt.figure(figsize=(13,7))
plt.title("Room Type on Neighbourhood Group")
sns.countplot(df.neighbourhood_group,hue=df.room_type, palette="muted")
plt.show()
```

# Plotting box plot of Neighbourhood Group VS Available Room
- Here we can see Available rooms in each Neighbourhood groups 
- Boxplot is reflecting the number of available rooms in neighbourhood groups 

```python
plt.style.use('classic')
plt.figure(figsize=(13,7))
plt.title("Neighbourhood Group vs. Available Room")
sns.boxplot(data=df, x='neighbourhood_group',y='availability_365',palette="dark")
plt.show()
```
